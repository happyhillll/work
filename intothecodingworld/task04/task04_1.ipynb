- txt 파일에서 문장들을 읽고, 형태소 기반 분석
    - txt 파일 내에는 한 줄에 한 문장이 들어있는 형태
    - 형태소 분석기 종류는 자유
- output 1
    - 특정 형태소를 포함하는 문장들을 dictionary에 모아둔다.
    - key : 형태소(예를 들어 학교/NNG), value : 해당 형태소를 포함하는 문장들
    - 예시
        
        {
        "회사/NNG": ["회사는 싫어.", "회사가 없어졌다.", …],
        
        “는/JX”: [”회사는 싫어”, “나는 집이 좋아”, …],
        
        …
        }
        
- output 2
    - txt 파일 문장 내에 형태소 빈도
    - 예시
        - [(”회사/NNG, 2”), (”는/JX”, 1), … ]  → txt 파일 내에 **회사/NNG**가 2번 발생, **는/JX**가 1번 발생

```python
import kiwipiepy  # kiwipiepy 모듈 import

kiwi = kiwipiepy.Kiwi()  # Kiwi 객체 생성

with open('newjeans.txt', 'r', encoding='utf-8') as f:  # text.txt 파일을 읽기 모드로 열기
    for line in f:  # 파일의 각 라인에 대해 반복
        sentence = line.strip()  # 라인에서 공백 제거 후 문장 변수에 저장
        result = kiwi.analyze(sentence)  # Kiwi 객체를 사용하여 문장 형태소 분석
        for word in result[0][0]:  # 분석 결과에서 첫번째 문장의 첫번째 형태소에 대해 반복
            print(word[0], word[1])  # 형태소와 품사 출력
```
